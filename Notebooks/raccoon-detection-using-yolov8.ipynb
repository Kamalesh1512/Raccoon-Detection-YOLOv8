```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')
```

```python
history=model.train(data='/kaggle/input/raccoon-images-dataset/Raccoon/data.yaml',
           imgsz=640,epochs=50,batch=32,name='yolov8n_v1')
```

## Model Validation

```python
metrics=model.val()
```

```python
metrics.box.map50
```
